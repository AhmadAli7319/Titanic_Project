## What is Data Exploration / EDA?

#### Data exploration (or Exploratory Data Analysis) is the process of understanding your data before building any model.
#### It’s like “getting to know your dataset” — what’s in it, how it behaves, and what patterns or problems might exist.

#### Key goals of data exploration:

``` Understand the structure of your data ```

- 1 Number of rows and columns

- 2 Data types (numeric, categorical, dates)

- 3 Missing values

``` Understand the distribution of variables```

- 1 Numeric columns: mean, median, min, max, quartiles, histograms

- 2 Categorical columns: value counts, frequency of each category

``` Check relationships between variables```

- 1 Correlations between numeric variables

- 2 How categorical variables relate to the target variable (Survived in Titanic)

- 3 Visualizations like scatter plots, box plots, bar plots

```Detect anomalies or outliers```

- 1 Values that don’t make sense or are extreme

- 2 Can affect your model if not handled

```Get insights for feature engineering```

- 1Combine columns, create new features

- 2Decide how to handle missing values, outliers, or categorical encoding

# **📘 Pandas Joins — Markdown Notes**
## **What is a Join?**

A **join** combines two DataFrames based on matching values in one or more columns.
Used when you have related data stored separately.

---

# **Types of Joins**

## **1️⃣ Inner Join**

* Returns only the rows where the **values match** in both DataFrames.
* Unmatched rows are removed.
* Most commonly used join.

**Code:**

```python
pd.merge(df1, df2, how='inner', on='ID')
```

**Meaning:**
Only keep data where `ID` exists in both `df1` and `df2`.

---

## **2️⃣ Left Join**

* Returns **all rows from the left DataFrame**.
* Adds matching rows from the right DataFrame.
* If a value in the right DataFrame does not match, the result shows `NaN`.

**Code:**

```python
pd.merge(df1, df2, how='left', on='ID')
```

**Meaning:**
Keep everything from `df1`; match data from `df2` if possible.

---

## **3️⃣ Right Join**

* Returns **all rows from the right DataFrame**.
* Adds matching rows from the left DataFrame.
* Non-matching values show `NaN`.

**Code:**

```python
pd.merge(df1, df2, how='right', on='ID')
```

**Meaning:**
Opposite of left join — keep everything from `df2`.

---

## **4️⃣ Outer Join (Full Join)**

* Returns **all rows from both DataFrames**.
* Keeps unmatched rows from both sides.
* Missing values are filled with `NaN`.

**Code:**

```python
pd.merge(df1, df2, how='outer', on='ID')
```

**Meaning:**
Keep all records; wherever a match is missing, fill with `NaN`.

---

## **5️⃣ Cross Join**

* Combines every row of the first DataFrame with every row of the second.
* Does **not** require a matching column.
* Produces a cartesian product.

**Code:**

```python
pd.merge(df1, df2, how='cross')
```

**Meaning:**
Every row from `df1` paired with every row from `df2`.

---

# **Important Notes**

* The join column must have matching **values**, not just matching **names**.
* You can use different column names from each DataFrame:

```python
pd.merge(df1, df2, left_on='A', right_on='B')
```

* Joins are very useful in data analysis, cleaning, EDA, and real-world datasets.
